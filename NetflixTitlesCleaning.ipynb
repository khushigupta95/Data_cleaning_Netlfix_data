# Data Cleaning and Manipulation


## Let's create a copy of dataset

```python
netflix = df.copy()
netflix.sample(5)
```

## Lets find if there are any Duplicated values

```python
netflix.duplicated().sum()
```

## Lets identify columns with missing values

```python
netflix.isna().sum().sort_values(ascending = False)
```

## Dealing with missing values

###  Directors

```python
no_director = netflix[netflix['director'].isnull()].index
no_director
```

```python
netflix.drop(no_director,axis = 0 , inplace = True)
netflix.isnull().sum()
```

## Dropping rows for columns with very few null values

### Rating

```python
netflix.dropna(subset=['rating'],axis = 0 , inplace = True)
```

```python
netflix.isnull().sum()
```

## Replacing missing values 

### Country

```python
netflix['country'] = netflix['country'].fillna("Unknown")
```

```python
netflix.isnull().sum()
```

```python
netflix.country.value_counts()
```

### Cast

```python
netflix['cast'] = netflix['cast'].fillna("No cast")
netflix.isnull().sum()
```

```python
netflix.cast.value_counts()
```

# Duration 

Duration is divided into two types - Seasons & Minutes .




Seasons associates with TV shows.

Minutes associates with Movies.

Lets combine both into a single data type .

```python
# dividing duration into Movies and TV shows
movies_df = netflix.loc[(netflix['type'] == 'Movie')]
movies_df.head(5)
```

```python
shows_df = netflix.loc[(netflix['type'] == 'TV Show')]
shows_df.head(5)
```

```python
movies_df.loc[:,'duration'] = movies_df['duration'].astype(str).str.replace(" min", "", regex=False)
movies_df.head(5)

```

```python
shows_df.loc[:,'duration'] = shows_df['duration'].astype(str).str.replace(" Season", "", regex=False)
shows_df.head(5)
```

```python
shows_df.loc[:,'duration'] = shows_df['duration'].astype(str).str.replace("s", "", regex=False)
shows_df.duration.sample(5)
```

```python
shows_df.duration.value_counts()
```
